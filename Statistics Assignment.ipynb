````python
#(ii) Develop a program to compute the mode of a list of integers.

from collections import Counter

def find_mode(numbers):
    # Use Counter to tally the occurrences of each number in the list
    count = Counter(numbers)
    
    # Determine the highest occurrence count
    highest_count = max(count.values())
    
    # Identify all numbers that occur with the highest frequency
    modes = [num for num, freq in count.items() if freq == highest_count]
    
    return modes
````

````python
# (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

def compute_weighted_average(values, weights):
    # Ensure that the number of values matches the number of weights and that weights are not empty
    if len(values) != len(weights) or not weights:
        return 0
    
    # Compute the total of each value multiplied by its corresponding weight
    total_weighted_sum = sum(v * w for v, w in zip(values, weights))
    
    # Compute the total sum of all weights
    total_weights = sum(weights)
    
    # Calculate the weighted average by dividing the total weighted sum by the sum of weights
    return total_weighted_sum / total_weights
````

````python
#  (iv) Write a Python function to find the geometric mean of a list of positive numbers.


import math

def find_geometric_mean(numbers):
    # Ensure all numbers are positive
    if any(n <= 0 for n in numbers):
        raise ValueError("Geometric mean can only be calculated for positive values")
    
    # Calculate the product of all the numbers in the list
    product = math.prod(numbers)
    
    # Compute the geometric mean by taking the root of the product
    return product ** (1 / len(numbers))
````

````python
 #(v) Create a program to calculate the harmonic mean of a list of values.
    
    def compute_harmonic_mean(numbers):
    # Verify that no number in the list is zero
    if any(n == 0 for n in numbers):
        raise ValueError("Harmonic mean cannot be calculated with zero values")
    
    # Sum up the reciprocals of each number in the list
    reciprocal_sum = sum(1 / n for n in numbers)
    
    # Calculate the harmonic mean by dividing the count of numbers by the reciprocal sum
    return len(numbers) / reciprocal_sum
````

````python
#(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum)

def find_midrange(numbers):
    # Determine the lowest value in the list
    lowest = min(numbers)
    
    # Determine the highest value in the list
    highest = max(numbers)
    
    # Compute the midrange by averaging the lowest and highest values
    return (lowest + highest) / 2
````

````python
#(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.


def compute_trimmed_mean(numbers, trim_percentage):
    # Check that the trim percentage is within the valid range (0 to 50)
    if not (0 <= trim_percentage < 50):
        raise ValueError("Trim percentage must be within the range of 0 to 50")
    
    # Sort the list of numbers in ascending order
    sorted_numbers = sorted(numbers)
    
    # Determine the number of values to exclude from each end of the list
    number_to_trim = int(len(sorted_numbers) * trim_percentage / 100)
    
    # Exclude the calculated number of values from both ends
    trimmed_numbers = sorted_numbers[number_to_trim:-number_to_trim]
    
    # Calculate and return the mean of the remaining values
    return calculate_mean(trimmed_numbers)
````

````python
  #(iii) Create a program to find the variance and standard deviation of a list of numbers.
    
    import numpy as np

    def compute_variance_and_std_dev(numbers):
    # Calculate the variance, which indicates the dispersion of the data points from the mean
    variance = np.var(numbers, ddof=0)
    
    # Calculate the standard deviation, the square root of the variance, showing the average deviation from the mean
    std_dev = np.std(numbers, ddof=0)
    
    # Return the computed variance and standard deviation
    return variance, std_dev
````

````python
#(iv) Implement a function to compute the interquartile range (IQR) of a list of values.

def calculate_iqr(numbers):
    # Determine the value at the 25th percentile (first quartile) of the dataset
    q1 = np.percentile(numbers, 25)
    
    # Determine the value at the 75th percentile (third quartile) of the dataset
    q3 = np.percentile(numbers, 75)
    
    # Find the interquartile range (IQR) by subtracting the first quartile value from the third quartile value
    return q3 - q1
````

````python
#(v) Build a program to calculate the coefficient of variation for a dataset.

def compute_coefficient_of_variation(numbers):
    # Compute the average (mean) of the data
    mean = np.mean(numbers)
    
    # Compute the standard deviation of the data
    std_dev = np.std(numbers, ddof=0)
    
    # Calculate the coefficient of variation as a percentage by dividing the standard deviation by the mean
    return (std_dev / mean) * 100
````

````python
# (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.

def compute_mean_absolute_deviation(numbers):
    # Find the average (mean) of the numbers
    mean = np.mean(numbers)
    
    # Determine the mean absolute deviation (MAD) by averaging the absolute differences from the mean
    mad = np.mean([abs(x - mean) for x in numbers])
    
    # Return the calculated mean absolute deviation
    return mad
````

````python
#(vii) Create a program to calculate the quartile deviation of a list of values.

def compute_quartile_deviation(numbers):
    # Find the value at the 25th percentile (first quartile) of the data
    q1 = np.percentile(numbers, 25)
    
    # Find the value at the 75th percentile (third quartile) of the data
    q3 = np.percentile(numbers, 75)
    
    # Calculate the quartile deviation by taking half of the range between the third and first quartiles
    return (q3 - q1) / 2
````

````python
#3.  Write a Python class representing a discrete random variable with methods to calculate its expected
#value and variance.

import numpy as np

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """Initialize the discrete random variable with a set of values and their associated probabilities.
        
        :param values: List of possible outcomes for the random variable.
        :param probabilities: List of probabilities corresponding to each value.
        """
        # Check that the number of values matches the number of probabilities
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities lists must be of the same length.")
        
        # Verify that the total of probabilities is close to 1
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("The total of probabilities must equal 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """Calculate the expected value (mean) of the discrete random variable.
        :return: The mean of the random variable.
        """
        # Calculate the mean by multiplying each value by its probability and summing the results
        return sum(v * p for v, p in zip(self.values, self.probabilities))
    
    def variance(self):
        """Compute the variance of the discrete random variable.
         :return: The variance of the random variable.
        """
        # Find the mean of the random variable
        mean = self.expected_value()
        
        # Calculate the variance by finding the average of the squared differences from the mean
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))

# Instantiate the DiscreteRandomVariable class with defined values and their probabilities
rv = DiscreteRandomVariable(values=[1, 2, 3], probabilities=[0.2, 0.5, 0.3])

# Print the expected value
print("Expected Value:", rv.expected_value())

# Print the variance
print("Variance:", rv.variance())
````

````python
#5. Create a Python 6unction to generate random samples 6rom a given probability distribution (e.g.,
#binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples_and_calculate_stats(distribution, params, num_samples):
    """Create random samples from a specified probability distribution and compute their mean and variance.

    param distribution: Specifies the type of distribution to sample from ('binomial' or 'poisson').
    param params: Parameters needed for the distribution.
        - For 'binomial': (n, p), where n is the number of trials and p is the success probability.
        - For 'poisson': lambda_, which represents the average rate of occurrences.

     param num_samples: The number of samples to generate.
    
    return: A tuple containing the mean and variance of the samples.
    """
    if distribution == 'binomial':
        # Extract the parameters for the binomial distribution
        n, p = params
        # Generate samples from a binomial distribution
        samples = np.random.binomial(n, p, size=num_samples)
    elif distribution == 'poisson':
        # Extract the parameter for the Poisson distribution
        lambda_ = params
        # Generate samples from a Poisson distribution
        samples = np.random.poisson(lambda_, size=num_samples)
    else:
        # Raise an exception if the distribution type is not recognized
        raise ValueError("Invalid distribution type. Use 'binomial' or 'poisson'.")

    # Compute the mean of the generated samples
    mean = np.mean(samples)
    # Compute the variance of the generated samples
    variance = np.var(samples)

    return mean, variance

# Example usage
# Create samples from a binomial distribution with 10 trials and a success probability of 0.5
mean, variance = generate_samples_and_calculate_stats('binomial', (10, 0.5), 1000)
print(f"Binomial Distribution - Mean: {mean:.2f}, Variance: {variance:.2f}")

# Create samples from a Poisson distribution with a mean rate of 5
mean, variance = generate_samples_and_calculate_stats('poisson', 5, 1000)
print(f"Poisson Distribution - Mean: {mean:.2f}, Variance: {variance:.2f}")
````

````python
#6. Write a Python script to generate random numbers 6rom a Gaussian (normal) distribution and compute
#the mean, variance, and standard deviation o6 the samples.


import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    """
    Generate random samples from a normal (Gaussian) distribution.
 
    param mean: The average value of the Gaussian distribution.
       param std_dev: The standard deviation, indicating the spread of the distribution.
     param num_samples: The number of samples to be generated.
        return: An array of samples drawn from the specified Gaussian distribution.
    """
    # Generate random data following a Gaussian distribution with the specified mean and standard deviation
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    return samples

def compute_statistics(samples):
    """Calculate the mean, variance, and standard deviation of the provided samples.
         An array containing the generated samples.
    A tuple with the mean, variance, and standard deviation of the samples.
    """
    # Compute the average value of the samples
    mean = np.mean(samples)
    
    # Determine how much the samples deviate from the mean (variance)
    variance = np.var(samples)
    
    # Calculate the standard deviation, which is the square root of the variance
    std_dev = np.std(samples)
    
    return mean, variance, std_dev

# Configuration for the Gaussian distribution
mean = 0          # The center of the distribution
std_dev = 1            # The spread or width of the distribution
num_samples = 1000  # Number of random samples to generate

# Generate samples from the Gaussian distribution
samples = generate_gaussian_samples(mean, std_dev, num_samples)

# Compute statistical measures of the samples
mean, variance, std_dev = compute_statistics(samples)

# Output the computed mean, variance, and standard deviation
print(f"Mean: {mean:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
````

````python
#(iii) Write a function that calculates the covariance between two columns.

import numpy as np

def calculate_covariance(column1, column2):
    """
    Calculate the covariance between two sets of numerical data.

    column1: A sequence of numerical values for the first dataset.
    column2: A sequence of numerical values for the second dataset.
    return: The covariance between the two datasets.
    """
    # Compute the covariance matrix for the two datasets
    # This matrix is 2x2, with the covariance values between the datasets in the off-diagonal elements
    covariance_matrix = np.cov(column1, column2, ddof=0)
    
    # Retrieve the covariance value from the matrix
    # It is found at position [0, 1] which represents the covariance between the first and second dataset
    return covariance_matrix[0, 1]
````

````python
#(iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.

import numpy as np

def calculate_pearson_correlation(column1, column2):
    """
    Calculate the Pearson correlation coefficient between two sets of numerical data.

    :param column1: Array-like structure of values for the first set of data.
    :param column2: Array-like structure of values for the second set of data.
    :return: The Pearson correlation coefficient between the two data sets.
    """
    # Compute the correlation coefficient matrix for the provided datasets
    # This matrix indicates the degree of linear relationship between the datasets
    correlation_matrix = np.corrcoef(column1, column2)
    
    # Retrieve the Pearson correlation coefficient from the matrix
    # The coefficient is located at [0, 1] in the matrix, reflecting the correlation between the two data sets
    return correlation_matrix[0, 1]
````

````python
11) A company wants to test iN a new website layout leads to a higher conversion rate (percentage oN visitors
who make a purchase). They collect data Nrom the old and new layouts to compare.

To generate the data use the following command:

```python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)

  ```

Apply z-test to find which layout is successful.
````

````python
import numpy as np
import scipy.stats as stats

# Create datasets for old and new website layouts
# Each array indicates whether a purchase was made (1) or not (0) from total visitors
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Determine the conversion rates for both layouts
# Conversion rate is calculated as the proportion of purchases
old_conversion_rate = np.mean(old_layout)
new_conversion_rate = np.mean(new_layout)

# Compute the number of purchases and the total number of visitors for each layout
old_successes = np.sum(old_layout)  # Purchases in the old layout
new_successes = np.sum(new_layout)  # Purchases in the new layout
old_total = len(old_layout)  # Visitors to the old layout
new_total = len(new_layout)  # Visitors to the new layout

# Calculate the combined proportion of purchases
# This represents the overall purchase rate across both layouts
pooled_proportion = (old_successes + new_successes) / (old_total + new_total)

# Determine the standard error of the difference between conversion rates
# This quantifies the expected variability in the difference
standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1 / old_total + 1 / new_total))

# Compute the Z-score
# The Z-score shows how many standard deviations the observed difference in conversion rates is from the expected value
z_score = (new_conversion_rate - old_conversion_rate) / standard_error

# Calculate the p-value
# The p-value represents the likelihood of observing a difference as extreme as this one, assuming no real effect
p_value = 1 - stats.norm.cdf(z_score)  # One-tailed test to assess if the new layout is more effective

# Output the results
print(f"Old Conversion Rate: {old_conversion_rate:.4f}")
print(f"New Conversion Rate: {new_conversion_rate:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Make a decision based on the significance level
alpha = 0.05  # The threshold for statistical significance
if p_value < alpha:
    print("The new layout significantly improves the conversion rate.")
else:
    print("There is no significant difference in conversion rates between the old and new layouts.")
````

````python
import numpy as np
import scipy.stats as stats

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    """
    Compare click-through rates between two website layouts using an A/B test.

    :param layout_a_clicks: List of click-through rates for layout A.
    :param layout_b_clicks: List of click-through rates for layout B.
    :return: t-statistic, degrees of freedom, and p-value.
    """
    # Convert lists to numpy arrays for easier calculations
    a = np.array(layout_a_clicks)
    b = np.array(layout_b_clicks)

    # Compute the average click-through rate for each layout
    mean_a = np.mean(a)
    mean_b = np.mean(b)

    # Compute the standard deviation for each layout
    std_a = np.std(a, ddof=1)  # Variability of layout A
    std_b = np.std(b, ddof=1)  # Variability of layout B

    # Calculate the sample sizes for both layouts
    n_a = len(a)
    n_b = len(b)

    # Determine the standard error of the difference between the two means
    pooled_std_error = np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Compute the t-statistic
    # This indicates how far the difference in average click-through rates is from zero
    t_statistic = (mean_a - mean_b) / pooled_std_error

    # Calculate the degrees of freedom for the test
    # This is used to find the right t-distribution for the p-value calculation
    df = ((std_a**2 / n_a + std_b**2 / n_b) ** 2) / \
         (((std_a**2 / n_a) ** 2 / (n_a - 1)) + ((std_b**2 / n_b) ** 2 / (n_b - 1)))

    # Compute the p-value
    # The p-value shows the probability of observing a difference at least as extreme as the one found if there was no real difference
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))  # Two-tailed test

    return t_statistic, df, p_value

# Example data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test
t_statistic, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Output the results
print(f"T-statistic: {t_statistic:.4f}")  # Shows how many standard deviations apart the means are
print(f"Degrees of Freedom: {df:.4f}")     # Reflects the sample sizes and variances in the test
print(f"P-value: {p_value:.4f}")           # Indicates the probability of the observed result under the null hypothesis

# Interpretation
alpha = 0.05  # Significance level to assess if the result is statistically significant
if p_value < alpha:
    print("There is a significant difference in click-through rates between the two layouts.")
else:
    print("There is no significant difference in click-through rates between the two layouts.")
````

````python
import numpy as np
import scipy.stats as stats

def compare_satisfaction(branch_a_scores, branch_b_scores):
    """
    Compare customer satisfaction scores from two different restaurant branches using a two-sample t-test
    to determine if there is a significant difference.

    :param branch_a_scores: Satisfaction scores for branch A.
    :param branch_b_scores: Satisfaction scores for branch B.
    :return: t-statistic, degrees of freedom, p-value.
    """
    # Convert the lists of scores into numpy arrays for easier calculations
    branch_a = np.array(branch_a_scores)
    branch_b = np.array(branch_b_scores)

    # Compute the mean satisfaction score for each branch
    mean_branch_a = np.mean(branch_a)
    mean_branch_b = np.mean(branch_b)

    # Compute the standard deviation for each branch
    # This shows how much the scores deviate from the average
    std_branch_a = np.std(branch_a, ddof=1)  # For branch A
    std_branch_b = np.std(branch_b, ddof=1)  # For branch B

    # Find out the number of scores for each branch
    n_branch_a = len(branch_a)
    n_branch_b = len(branch_b)

    # Calculate the standard error of the difference between the means
    # This indicates how much variability there is in the difference of means
    pooled_std_error = np.sqrt((std_branch_a**2 / n_branch_a) + (std_branch_b**2 / n_branch_b))

    # Compute the t-statistic
    # This measures how far the difference in means is from zero, relative to the variability
    t_statistic = (mean_branch_a - mean_branch_b) / pooled_std_error

    # Calculate the degrees of freedom
    # This helps to select the appropriate t-distribution for the test
    df = ((std_branch_a**2 / n_branch_a + std_branch_b**2 / n_branch_b) ** 2) / \
         (((std_branch_a**2 / n_branch_a) ** 2 / (n_branch_a - 1)) + ((std_branch_b**2 / n_branch_b) ** 2 / (n_branch_b - 1)))

    # Calculate the p-value
    # This gives the probability of seeing a difference as extreme as observed, assuming no real difference exists
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))  # Two-tailed test

    return t_statistic, df, p_value

# Sample satisfaction scores for the two branches
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Execute the t-test to compare the satisfaction scores
t_statistic, df, p_value = compare_satisfaction(branch_a_scores, branch_b_scores)

# Display the results
print(f"T-statistic: {t_statistic:.4f}")  # Reflects the difference in average scores adjusted for variability
print(f"Degrees of Freedom: {df:.4f}")     # Necessary for determining the appropriate t-distribution
print(f"P-value: {p_value:.4f}")           # Shows the likelihood of observing a difference as large or larger by chance

# Interpret the outcome
alpha = 0.05  # Significance level for determining if the result is statistically significant
if p_value < alpha:
    print("There is a statistically significant difference in customer satisfaction between the two branches.")
else:
    print("There is no statistically significant difference in customer satisfaction between the two branches.")
````

````python
21.  A political analyst wants to determine i@ there is a signi@icant association between age groups and voter
pre@erences FCandidate A or Candidate B). They collect data @rom a sample o@ 500 voters and classi@y
them into di@@erent age groups and candidate pre@erences. Per@orm a Chi-Square test to determine i@
there is a signi@icant association between age groups and voter pre@erences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
````

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create the dataset
# Set a random seed to ensure reproducibility
np.random.seed(0)

# Generate age group data for 500 voters
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)

# Generate voter preference data for 500 voters (either Candidate A or B)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Step 2: Build a contingency table
# Combine the age group and voter preference data into a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a table showing the count of each age group and voter preference combination
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Step 3: Conduct the Chi-Square test
# Perform the Chi-Square test to get the statistic, p-value, degrees of freedom, and expected frequencies
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat}")  # The Chi-Square test statistic value
print(f"P-Value: {p_val}")  # Indicates the statistical significance of the test
print(f"Degrees of Freedom: {dof}")  # Degrees of freedom for interpreting the statistic
print(f"Expected Frequencies Table:\n{ex}")  # Shows the expected frequencies if there were no association

# Interpretation of results
alpha = 0.05  # Standard significance threshold
# Compare the p-value to the significance level
if p_val < alpha:
    print("Reject the null hypothesis: There is a statistically significant relationship between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is no statistically significant relationship between age groups and voter preferences.")
````

````python
22. A company conducted a customer satisfaction survey to determine if there is a significant relationship
between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are
located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a ChiSquare test to determine if there is a significant relationship between product satisfaction levels and
customer regions.


Sample data:

```python

#Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)

data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])
````

````python
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Rows represent levels of product satisfaction (Satisfied, Neutral, Dissatisfied)
# Columns represent different customer regions (East, West, North, South)
data = np.array([[50, 30, 40, 20],  # Satisfied
                 [30, 40, 30, 50],  # Neutral
                 [20, 30, 40, 30]]) # Dissatisfied

# Step 1: Conduct the Chi-Square test
# Perform the Chi-Square test on the contingency table to compute the test statistic, p-value, degrees of freedom, and expected frequencies
chi2_stat, p_val, dof, ex = chi2_contingency(data)

# Display the results from the Chi-Square test
print(f"Chi-Square Statistic: {chi2_stat}")  # This statistic measures how much the observed data differs from what was expected
print(f"P-Value: {p_val}")  # The p-value indicates the likelihood that the observed differences are due to random chance
print(f"Degrees of Freedom: {dof}")  # The degrees of freedom are used to interpret the Chi-Square statistic
print(f"Expected Frequencies Table:\n{ex}")  # This table shows the frequencies we would expect if there were no association between satisfaction levels and regions

# Interpretation of results
alpha = 0.05  # The threshold for statistical significance
# Determine if the p-value is below the significance level to assess whether a significant relationship exists
if p_val < alpha:
    print("Reject the null hypothesis: A significant relationship exists between product satisfaction levels and customer regions.")
else:
    print("Fail to reject the null hypothesis: No significant relationship exists between product satisfaction levels and customer regions.")
````

````python
 23. A company implemented an employee training program to improve job performance (Effective, Neutral,
Ineffective). After the training, they collected data from a sample of employees and classified them based
on their job performance before and after the training. Perform a Chi-Square test to determine if there is a
significant difference between job performance levels before and after the training.


Sample data:

```python

# Sample data: Job performance levels before (rows) and after (columns) training

data = np.array([[50, 30, 20], [30, 40, 30], [20, 30, 40]])
````

````python
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Rows denote job performance levels before training (Effective, Neutral, Ineffective)
# Columns denote job performance levels after training (Effective, Neutral, Ineffective)
data = np.array([[50, 30, 20],  # Performance levels before training: Effective, Neutral, Ineffective
                 [30, 40, 30],  # Performance levels before training: Effective, Neutral, Ineffective
                 [20, 30, 40]]) # Performance levels before training: Effective, Neutral, Ineffective

# Step 1: Conduct the Chi-Square test
# This test assesses whether there is a notable difference between job performance levels before and after the training
chi2_stat, p_val, dof, ex = chi2_contingency(data)

# Display the results of the Chi-Square test
print(f"Chi-Square Statistic: {chi2_stat}")  # Reflects how much the actual performance data deviates from expected values under no change
print(f"P-Value: {p_val}")  # Shows the probability that the observed deviations are due to random variation
print(f"Degrees of Freedom: {dof}")  # Provides context for interpreting the Chi-Square statistic
print(f"Expected Frequencies Table:\n{ex}")  # Illustrates the expected counts if there were no effect of training on performance levels

# Interpretation of the results
alpha = 0.05  # The significance level used to determine statistical significance
# Check if the p-value is below the significance level to determine if there is a meaningful difference
if p_val < alpha:
    print("Reject the null hypothesis: A significant difference exists in job performance levels before and after the training.")
else:
    print("Fail to reject the null hypothesis: No significant difference is observed in job performance levels before and after the training.")
````

````python
 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The
company wants to determine if there is a significant difference in customer satisfaction scores among the
three product versions. They conducted a survey and collected customer satisfaction scores for each
version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
difference in customer satisfaction scores.


  Use the following data:

  ```python

  # Sample data: Customer satisfaction scores for each product version

  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

  premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

  deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

````

````python
import numpy as np
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product variant
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Step 1: Conduct the ANOVA test
# This test evaluates whether there are significant differences in satisfaction scores among the three product versions
f_stat, p_val = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Output the results of the ANOVA test
print(f"F-Statistic: {f_stat}")  # Shows the ratio of variance between groups to variance within groups
print(f"P-Value: {p_val}")  # Reflects the probability that the observed differences are due to chance

# Interpreting the results
alpha = 0.05  # The threshold for determining statistical significance
# Assess if the p-value is below the significance level to determine if there are significant differences
if p_val < alpha:
    print("Reject the null hypothesis: Significant differences exist in customer satisfaction scores among the product variants.")
else:
    print("Fail to reject the null hypothesis: No significant differences are found in customer satisfaction scores among the product variants.")
````
