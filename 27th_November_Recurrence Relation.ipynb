```python
# 1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)= 5.

def recurrence_relation(n):
  if n == 0:
    return 5
  else:
    return 3 * recurrence_relation(n-1) + 12 * n

# calculate T(2)
result_T2 = recurrence_relation(2)
print(result_T2)
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:
# a. T(n) = T(n-1) + c

T(n) = T(n-1) + c

T(n-1) = T((n-1)-1) + c

T(n) = T(n-2) + 2c

T(n-2) = T((n-2)-1) + c

T(n) = T(n-3) + 3c

T(n) = T(n-k) + k.c

# This recurrsion continues till n-k = 1, which implies k = n-1

T(n) = T(n-(n-1)) + (n-1).c

T(n) = T(1) + (n-1).c

T(n) = 1 + (n-1).c
```

```python
# b. T(n) = 2T(n/2) + n
 
T(n) = 2T(n/2) + n..............................1st equation

T(n/2) = 2T(n/2^2) + n/2
T(n) = 2[2T(n/2^2) + n/2] + n

T(n) = 2^2T(n/2^2) + n + n .....................2nd equation

T(n/2^2) = 2T(n/2^3) + n/2^2
T(n) = 2^2[2T(n/2^3) + n/2^2] + n + n
T(n) = 2^3T(n/2^3) + n + n + n...................3rd equation


T(n) = 2^kT(n/2^k) + k.n

# n/2^k = 1, therefore T(1) = 1
# n = 2^k 
# log₂n = k log₂2
# k = log₂n

T(n) = 2^k + k.n
T(n) = n + nlog₂n
```

```python
# c. T(n) = 2T(n/2) + c

T(n) = 2T(n/2) + c.......................1st equation

T(n/2) = 2T(n/2^2) + c
T(n) = 2[2T(n/2^2) + c] + c

T(n) = 2^2T(n/2^2) + 2c + c..............2nd equation 

T(n/2^2) = 2T(n/2^3) + c
T(n) = 2^2[2T(n/2^3) + c] + 2c + c

T(n) = 2^3T(n/2^3) + 2^2c + 2c + c.......3rd equation

T(n) = 2^kT(n/2^k) + 2^k-1.c + 2^k-2.c........+ 2c + c

# This recurrsion continues till n/2^k = 1, which implies k = log₂n

T(n) = n + c(2^log₂n-1 + 2^log₂n-2+........+ 2 +1)

T(n) = n + c(n-1)
```

```python
# d. T(n) = T(n/2) + c

T(n) = T(n/2) + c...................1st equation

T(n/2) = T(n/2^2) + c

T(n) = T(n/2^2) + c + c..............2nd equation

T(n/2^2) = T(n/2^3) + c

T(n) = T(n/2^3) + c + c

T(n) = T(n/2^k) + k.c

# n/2^k = 1, therefore T(1) = 1
# n = 2^k 
# log₂n = k log₂2
# k = log₂n

T(n) = 1 + clog₂n
```
